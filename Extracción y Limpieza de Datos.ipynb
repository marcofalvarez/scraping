```python
"""
This script contains two functions to scrape salary data from two different websites: capology.com and footystats.org.
The function `obtener_salarios` scrapes salary data from capology.com and returns a pandas DataFrame with the data.
The function `obtener_salarios` scrapes salary data from footystats.org and returns a pandas DataFrame with the data.
"""

# Diccionario que mapea nombres de países a sus códigos de dos letras en capology.com
map_pais = {
    'Spain': 'es',
    'Germany': 'de',
    'France': 'fr',
    'Italy': 'it',
    'England': 'uk'
}

# Diccionario que mapea nombres de ligas a sus identificadores en capology.com
map_liga = {
    'La Liga': 'la-liga',
    'Bundesliga': '1-bundesliga',
    'Ligue 1': 'ligue-1',
    'Serie A': 'serie-a',
    'Premier League': 'premier-league'
}

def quitar_euro(x):
    return float(x.replace('€', ''))

def get_salary_data(country, league, season=None):
    if season and season != '2023-2024':
        return requests.get(f'https://www.capology.com/{map_pais[country]}/{map_liga[league]}/salarios/{season}')
    else:
        return requests.get(f'https://www.capology.com/{map_pais[country]}/{map_liga[league]}/salarios')

def obtener_salarios(pais, liga, temporada=None):
    if temporada == None:
        temporada = '2023-2024'
    response = get_salary_data(pais, liga, temporada)
    html = response.content
    # Crear un objeto BeautifulSoup con el HTML
    sopa = BeautifulSoup(html, "html.parser")
    scripts = sopa.find_all('script')
    ds = str(scripts[-1].next)
    psd = parse(ds)
    body = psd['body']
    # Lista para almacenar los diccionarios con la información de cada jugador
    dicts = []
    # Iterar sobre los elementos del cuerpo del script
    for belem in body:
        # Si el elemento es una declaración de variable
        if belem['type'] == 'VariableDeclaration':
            # Obtener la variable declarada
            var_decl = belem['declarations'][0]
            # Si la variable es 'data'
            if(var_decl['type'] == 'VariableDeclarator' and
                var_decl['id']['name'] == 'data' and
                var_decl['id']['type'] == 'Identifier'):
                # Obtener los elementos de la lista asignada a la variable 'data'
                elems = var_decl['init']['elements']
                # Iterar sobre los elementos
                for elem in elems:
                    # Obtener las propiedades del elemento
                    props = elem['properties']
                    # Crear un diccionario para almacenar los valores de las propiedades
                    result_dict = {}
                    # Iterar sobre las propiedades
                    for prop in props:
                        # Obtener el nombre de la propiedad
                        key = prop['key']['value']
                        # Obtener el valor de la propiedad
                        value = get_value(prop['value'])
                        # Agregar la propiedad y su valor al diccionario
                        result_dict[key] = value
                    # Agregar el diccionario a la lista de diccionarios
                    dicts.append(result_dict)
        else:
            pass
    # Lista para almacenar los diccionarios con la información de cada jugador, con los nombres de las columnas en español
    dimproved = []

    # Iterar sobre los diccionarios que contienen la información de cada jugador
    for dich in dicts:
        # Crear un nuevo diccionario con las claves y valores mejorados
        dik = {'name': BeautifulSoup(dich.get('name'), "html.parser").text, # Obtener el nombre del jugador y eliminar las etiquetas HTML
               'club': BeautifulSoup(dich.get('club'), "html.parser").text, # Obtener el nombre del club y eliminar las etiquetas HTML
               'verified': BeautifulSoup(dich.get('verified'), "html.parser").contents[0]['src'].endswith(
                   "verified-green.svg") if dich.get('verified') is not None else None, # Obtener el estado de verificación del jugador y convertirlo a un valor booleano
               'years': eval(dich.get('years')) if dich.get('years') is not None else None, # Obtener el número de años restantes en el contrato del jugador y convertirlo a un valor numérico
               'active': eval(dich.get('active')) if dich.get('active') is not None else None, # Obtener el estado de actividad del jugador y convertirlo a un valor booleano
               'loan': eval(dich.get('loan')) if dich.get('loan') is not None else None, # Obtener el estado de préstamo del jugador y convertirlo a un valor booleano
               'starter': dich.get('status') == 'Starter' if dich.get('status') is not None else None, # Obtener el estado de titularidad del jugador y convertirlo a un valor booleano
               'signed': dich.get('signed'), # Obtener la fecha de firma del contrato del jugador
               'expiration': dich.get('expiration'), # Obtener la fecha de vencimiento del contrato del jugador
               'position': dich.get('position'), # Obtener la posición del jugador
               'position_detail': dich.get('position_detail'), # Obtener detalles adicionales sobre la posición del jugador
               'age': dich.get('age'), # Obtener la edad del jugador
               'country': dich.get('country')} # Obtener el país de origen del jugador

        # Iterar sobre las claves que contienen información sobre los salarios del jugador
        for key in ['weekly_gross_eur', 'annual_gross_eur', 'adjusted_gross_eur', 'weekly_net_eur',
                'annual_net_eur', 'adjusted_net_eur', 'bonus_gross_eur', 'bonus_net_eur',
                'total_gross_eur', 'total_net_eur', 'release_eur']:
            # Obtener el valor de la clave y convertirlo a un valor numérico, eliminando el símbolo de euro si está presente
            dik[key[:-len("_eur")]] = quitar_euro(dich.get(key)) if (dich.get(key) not in [None, 0 ,'']) else None

        # Agregar el diccionario mejorado a la lista de diccionarios
        dimproved.append(dik)

    # Crear un DataFrame a partir de la lista de diccionarios mejorados
    df = pd.DataFrame(dimproved)



    # Renombrar las columnas utilizando el diccionario de mapeo
    # Renombrar las columnas utilizando el diccionario de mapeo
    df = df.rename(columns=column_name_mapping)

    # Eliminar columnas redundantes (linealmente dependientes de las anuales)
    df.drop(['bruto_semanal', 'neto_semanal'], axis=1, inplace=True)
    # En este solo hay valores nulos, así que lo quitamos también
    df.drop(['bruto_ajustado', 'neto_ajustado'], axis=1, inplace=True)

    # Convertir las fechas de firma y vencimiento del contrato a objetos datetime.date
    df['firma_contrato'] = pd.to_datetime(df['firma_contrato']).dt.date
    df['vencimiento_contrato'] = pd.to_datetime(df['vencimiento_contrato']).dt.date

    df['posicion'] = df['posicion'].map(pos)

    df.drop('posicion_detallada', axis=1, inplace=True)

    df['pais'] = pais
    df['liga'] = liga
    df['temporada'] = temporada
    print(df)
    return df

def get_value(prop_value):
    """
    Esta función toma un diccionario como entrada y devuelve un valor basado en el tipo del diccionario.
    Si el diccionario contiene una clave 'value', devuelve el valor de esa clave.
    Si el diccionario contiene una clave 'type' con valor 'CallExpression', verifica la propiedad callee.
    Si la propiedad callee tiene una propiedad 'name' con valor 'formatMoney', llama a la función parse_currency_val.
    Si la propiedad callee tiene una propiedad 'name' con valor 'round' y el objeto callee tiene un nombre 'Math', redondea el valor del primer argumento.
    Si el objeto callee tiene una propiedad 'callee' con un nombre 'moment', devuelve el valor del primer argumento.
    Si no se cumplen ninguna de las condiciones anteriores, devuelve None.
    """
    if 'value' in prop_value:
        return prop_value['value']
    else:
        if prop_value['type'] == 'CallExpression':
            callee_ = prop_value['callee']
            if callee_['property']['name'] == 'formatMoney':
                return parse_currency_val(prop_value)
            else:
                callee_obj = callee_['object']
                if callee_obj.get('name') == 'Math' and \
                        callee_['property']['name'] == 'round':
                    return round(eval(prop_value['arguments'][0]['value']))
                elif callee_obj['callee']['name'] == 'moment':
                    return callee_obj['arguments'][0]['value']
                else:
                    pass
        else:
            pass


def parse_currency_val(prop_value):
    """
    Toma un objeto de propiedad y devuelve una cadena que representa su valor de moneda.

    Args:
        prop_value (dict): Un objeto de propiedad que contiene información sobre el valor de la moneda.

    Returns:
        str: Una cadena que representa el valor de la moneda.
    """
    raw_val = prop_value['arguments'][0]
    if raw_val['type'] == 'BinaryExpression':
        imm_value = eval(
            f"{raw_val['left']['value']}{raw_val['operator']}{raw_val['right']['value']}")
    elif raw_val['type'] == 'Literal':
        imm_value = eval(raw_val['value'])
    else:
        pass
    return str(imm_value) + prop_value['arguments'][1]['value'].strip()


k = pd.read_csv('./Salarios_Capology.csv')

pack_salarios = pd.concat([(obtener_salarios("Spain", "La Liga", '2021-2022')),
                           (obtener_salarios("Spain", "La Liga", '2022-2023')),
                           (obtener_salarios("Spain", "La Liga", '2023-2024')),
                           (obtener_salarios("England", "Premier League", '2021-2022')),
                           (obtener_salarios("England", "Premier League", '2022-2023')),
                           (obtener_salarios("England", "Premier League", '2023-2024')),
                           (obtener_salarios("Germany", "Bundesliga", '2021-2022')),
                           (obtener_salarios("Germany", "Bundesliga", '2022-2023')),
                           (obtener_salarios("Germany", "Bundesliga", '2023-2024')),
                           (obtener_salarios("Italy", "Serie A", '2021-2022')),
                           (obtener_salarios("Italy", "Serie A", '2022-2023')),
                           (obtener_salarios("Italy", "Serie A", '2023-2024')),
                           (obtener_salarios("France", "Ligue 1", '2021-2022')),
                           (obtener_salarios("France", "Ligue 1", '2022-2023')),
                           (obtener_salarios("France", "Ligue 1", '2023-2024'))],
                          ignore_index=True)
pack_salarios.to_csv("./Salarios_Capology_javier.csv", index=False)
```

```python
def rcsv(tempo, liga, type):
    return pd.read_csv('detail_%s/%s_%s_%s.csv' % (tempo, liga, type, tempo))

# Función para normalizar los nombres del equipo con regex
def normalize_team_name(name):
    pattern = re.compile(r'^\d+\.\s(.*)$')
    match = pattern.match(name)
    if match:
        return match.group(1)
    else:
        return name

def method_name(temporada, n_liga):
    df1 = rcsv(temporada, n_liga, 'actionzones')
    df2 = rcsv(temporada, n_liga, 'attackside')
    df3 = rcsv(temporada, n_liga, 'defensive')
    df3 = df3.rename(columns={'Shots pg': 'Shots pg Defensive'})
    df4 = rcsv(temporada, n_liga, 'detailed')
    df5 = rcsv(temporada, n_liga, 'goaltypes')
    df6 = rcsv(temporada, n_liga, 'offensive')
    df6 = df6.rename(columns={'Shots pg': 'Shots pg Offensive'})
    # Normalizar la columna "nombre del equipo" en cada DataFrame utilizando regex
    df_list = normalize_teams(df1, df2, df3, df4, df5, df6)
    # Combinar los DataFrames en función de la columna "nombre del equipo"
    combined_df = df1
    for df in df_list[1:]:
        combined_df = pd.merge(combined_df, df, on='Team', how='outer')
    combined_df['Temporada'] = temporada
    combined_df['Liga'] = n_liga
    return combined_df


def normalize_teams(*args):
    df_list = [df for df in args if hasattr(df, 'Team')]
    for df in df_list:
        try:
            df['Team'] = df['Team'].apply(normalize_team_name)
        except:
            pass
    return df_list


# Definir la función para convertir porcentajes usando regex
def method_name2(df, col_name):
    df[col_name] = df[col_name].apply(lambda x: float(re.compile(r'^(\d+)\s*(%)$').match(x).group(1)) / 100)
    return df


def method_name3(tempo):
    df1 = method_name(tempo, 'Bundesliga')
    df2 = method_name(tempo, 'LaLiga')
    df3 = method_name(tempo, 'Ligue 1')
    df4 = method_name(tempo, 'Premier League')
    df5 = method_name(tempo, 'Serie A')
    # Aplicar la función a la columna "Porcentaje" del DataFrame
    df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
    df = method_name2(df, 'Own Third')
    df = method_name2(df, 'Middle Third')
    df = method_name2(df, 'Opposition Third')
    df = method_name2(df, 'Left Side')
    df = method_name2(df, 'Middle of the pitch')
    df = method_name2(df, 'Right Side')

    return df

def summary(tempo, liga):
    csv = pd.read_csv('summary_%s/%s_summary_%s.csv' % (tempo, liga, tempo))
    csv['Liga'] = liga
    return csv

def method_name4(temporada):
    df1 = summary(temporada, 'Bundesliga')
    df2 = summary(temporada, 'LaLiga')
    df3 = summary(temporada, 'Ligue 1')
    df4 = summary(temporada, 'Premier League')
    df5 = summary(temporada, 'Serie A')

    # Normalizar la columna "nombre del equipo" en cada DataFrame utilizando regex
    df_list = normalize_teams(df1, df2, df3, df4, df5)
    # Combinar los DataFrames en función de la columna "nombre del equipo"

    combined_df = pd.concat(df_list, ignore_index=True)
    combined_df['Temporada'] = temporada
    return combined_df


def get_table(liga, tempo):
    return pd.read_csv('tables/%s_league_%s.csv' % (liga, tempo))

def method_name52(liga, tempo):
    t1 = get_table(liga, tempo)
    t1['Temporada'] = tempo
    return t1
def method_name5(liga):
    combined_df = pd.concat([(method_name52(liga, '21_22')), (method_name52(liga, '22_23')),
                             (method_name52(liga, '23_24'))], ignore_index=True)
    combined_df['Liga'] = liga
    return combined_df
```

```python
def normalize_step(cadena):
    cadena = cadena.strip()
    cadena = unidecode(cadena)
    if 'Milan' not in cadena and 'Milano' not in cadena:
        regex_pattern = r'\b(?:[A-Z]{2,3}|' + '|'.join(
            map(re.escape, ['VfB', 'VfL', 'SpVgg'])) + r')\b'  # Siglas de clubs
        cadena = re.sub(regex_pattern, '', cadena).strip()
    cadena = re.sub(r'\b\d+(?!\s*\.)', '', cadena).strip() #Para 09 Borussia Dortmund, Bayer 04 Leverkusen, etc.; Para años: 1899 Hoffenheim
    cadena = re.sub(r'\s+de\s+Futbol\s*$', '', cadena).strip() # Aplicar la sustitución solo si 'de Futbol' está al final
    cadena = re.sub(r'\s+Calcio\s*$', '', cadena).strip()
    cadena = re.sub(r'\s{2,}', ' ', cadena).strip() # Eliminar espacios redundantes
    match = re.search(r'\d+\s*\.\s*(.*)$', cadena)
    if match:
        cadena = match.group(1).strip()
    else:
        cadena = cadena.strip()



    if cadena in mapa_normalizacion:
        cadena = mapa_normalizacion[cadena]

    return cadena
```

```python
def ver_discrepancias(a, b):
    a = sorted(a)
    b = sorted(b)

    for elemento1, elemento2 in zip(a, b):
        if elemento1 != elemento2:
            print(f"Los elementos {elemento1} y {elemento2} no son iguales.")


def identificar_similitudes(teams):
    # Identificar similitudes entre los nombres de los equipos
    similarities = []
    for i in range(len(teams)):
        for j in range(i + 1, len(teams)):
            ratio = difflib.SequenceMatcher(None, teams[i], teams[j]).ratio()
            if ratio > 0.6:  # Puedes ajustar este umbral según tus necesidades
                similarities.append((teams[i], teams[j], ratio))
    # Imprimir las similitudes entre los nombres de los equipos
    for sim in similarities:
        print(f"Similaridad entre {sim[0]} y {sim[1]}: {sim[2]}")
```
