```python
import numpy as np
import pandas as pd

import requests
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("limpios/salarios_detail_por_jugador.csv")
df
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

sns.barplot(x = "liga", y = "bruto_anual", hue = "temporada", data = df, palette = "rainbow")
plt.show()
```

**Observamos los salarios que hay en cada liga en los ultimos años**

```python
fig, ax = plt.subplots(figsize = (10, 6))

sns.barplot(x = "liga", y = "bruto_anual", hue = "posicion", data = df, palette = "rainbow")
plt.show()
```

**En este gráfico podemos observar la diferencia de salarios que hay en cada liga según la posicion de los jugadores**

```python
# Creamos un pivot table para preparar los datos para el mapa de calor
heatmap_data = df.pivot_table(index='posicion', values='neto_anual', aggfunc='mean')

# Creamos el mapa de calor
plt.figure(figsize=(10, 6))
sns.heatmap(data=heatmap_data, annot=True, cmap='YlGnBu', fmt='.0f')
plt.title('Ingresos netos anuales por posición')
plt.show()
```

**Al igual que en el gráfico anterior, aquí podemos observar los ingresos netos anuales que obtienen los jugadores de las 5 grandes ligas en funcion de su posicion, donde se observa que los jugadores que más ingresos obtienen son los delanteros y centrocampistas**

```python
# Crea el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(df['edad'], df['neto_anual'], color='skyblue')
plt.title('Comparación de salario anual por rango de edad')
plt.xlabel('Edad')
plt.ylabel('Salario anual')
plt.savefig("bruto_cumple.png")
plt.show()
```

**Aquí podemos observar que lo normal en jugadores es que su "prime" en cuanto a ingresos se concentra en el rango de edad desde los 25 hasta los 35 años, aunque tambien podemos observar que las superestrellas salen de lo normal desde bien jovencitos. Un dato a destacar es que observamos que la antigüedad tambien se paga ya que los pocos jugadores que permanecen jugando desdepues de los 32 años siguen ganando un buen sueldo a pesar de no estar fisicamente igual que un jugador con una edad imferior**

```python
df3 = pd.read_csv("limpios/datos_por_equipo_1.csv")
df3
```

```python
# Filtramos los datos para la temporada 2022 y los 15 equipos con más goles a favor
df_2022 = df3[df3['temporada'] == 2022].nlargest(20, 'goles_a_favor')

# Creamos el gráfico de barras
plt.figure(figsize=(18, 8))
plt.bar(df_2022['equipo'], df_2022['goles_a_favor'], color='lightblue')
plt.title('Top 20 equipos con más goles a favor en la temporada 2022')
plt.xlabel('Equipo')
plt.ylabel('Goles a favor')
plt.xticks(rotation=45)
plt.savefig("equipo_goles.png")
plt.show()
```

**En este gráfico de barras podemos observar cuales fueron los equipos que más goles anotaron durante la pasada temporada y de ahí podemos sacar algunas conclusiones como que por ejemplo el Manchester City fuera el equipo que ganará todo y mayor superioridad mostrara respecto a otros equipos siendo el equipo que más goles anotó. En conclusión, una maquina de hacer goles**

```python
# Filtramos los datos para la temporada 2022
df_2022 = df3[df3['temporada'] == 2022]

# Agrupamos los datos por liga y sumamos los goles a favor
df_agrupado = df_2022.groupby('liga')['goles_a_favor'].sum()

colores = ['skyblue', 'lightgreen', 'lightcoral', 'lightpink', 'yellow'] 

# Creamos el gráfico de barras
plt.figure(figsize=(10, 6))
df_agrupado.plot(kind='bar', color=colores)
plt.title('Goles a favor por liga en la temporada 2022')
plt.xlabel('Liga')
plt.ylabel('Goles a favor totales')
plt.xticks(rotation=45)
plt.savefig("liga_goles.png")
plt.show()
```

**En este gráfico podemos observar las ligas donde más goles a favor hubo, lo cual es un dato importante ya que consideramos que a mayor goles, mayor es el espectaculo, entonces solo tienes que observar en que liga hay más goles, poner la tele, bsucar partidos de esa liga y a disfrutar**

```python
df_2022 = df3[df3['temporada'] == 2022]  # Filtramos por la temporada de 2002

plt.figure(figsize=(12, 60))  # Definimos el tamaño de la figura

# Creamos el hisplot horizontal con Seaborn
sns.histplot(data=df_2022, y='equipo', hue='liga', weights='goles_a_favor', multiple='stack')

# Añadimos título y etiquetas
plt.title('Goles a favor por liga y equipo durante la temporada de 2022')
plt.xlabel('Goles a favor')
plt.ylabel('Equipo')

# Mostramos la visualización
plt.show()
```
